```python
import re

def clean_text_function(text):
    if isinstance(text, str):  # Check if the value is a string
        cleaned_text = text.lower()  # Convert text to lowercase
        cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)  # Remove special characters
        cleaned_text = re.sub(r'\d+', '', cleaned_text)  # Remove numbers
        cleaned_text = cleaned_text.strip()  # Remove leading/trailing spaces
        return cleaned_text
    else:
        return ''  # Return an empty string for non-string values

df['clean_text'] = df['reviews.text'].apply(clean_text_function)
```

```python


# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load the dataset
# Assuming your dataset is stored in a pandas DataFrame called 'df'

# Text cleaning and normalization
df['clean_text'] = df['reviews.text'].apply(lambda x: clean_text_function(x))






```

```python

# Text cleaning and normalization
df['clean_text'] = df['reviews.text'].apply(lambda x: clean_text_function(x))
```

```python
# Tokenization
df['tokens'] = df['clean_text'].apply(lambda x: nltk.word_tokenize(x))




```

```python
# Stop words removal
stop_words = set(stopwords.words('english'))
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word.lower() not in stop_words])
df['tokens']
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['tokens'] = df['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df['tokens']
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
df['tokens'] = df['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df['tokens']
```

```python
# Stemming
stemmer = PorterStemmer()
df['tokens'] = df['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])
```
